```python
def compute_total(row, col1, col2):
    val = (row[col1] if row[col1] != -1 else 0) + (row[col2] if row[col2] != -1 else 0)
    return val
```

```python
pls_ae_data_map = {}
for year in range(2007, 2021):
    for _, row in pls_ae_data[year].iterrows():
        if not row['FSCSKEY'] in pls_ae_data_map.keys():
            pls_ae_data_map[row['FSCSKEY']] = {'id': row['FSCSKEY']}
        pls_ae_data_map[row['FSCSKEY']][year] = {}
        if year == 2020:
            for col in ae_per_cols:
                pls_ae_data_map[row['FSCSKEY']][col] = row[col]
            for col in ae_rel_cols:
                if col in dir(row):
                    pls_ae_data_map[row['FSCSKEY']][year][col] = row[col]
        else:
            for col in ae_other_cols:
                if col in dir(row):
                    pls_ae_data_map[row['FSCSKEY']][year][col] = row[col]
```

```python
edu_att_2020 = pd.read_csv('Education_Attainment/ACS_Education_Attainment_2020_state.csv')
```

```python
def percent(s):
    if '(X)' in s:
        return 0
    return round(float(s[:-1]), 1)
```

```python
state_edu_2020_map = {}
age_groups = ['18-24', '25-34', '35-44', '45-64', '65+', '25+']
for state in STATE_ABBR_MAP.keys():
    state_edu_2020_map[state] = {}
    for age_group in age_groups:
        state_edu_2020_map[state][age_group] = {}
        state_edu_2020_map[state][age_group]['Bachelor and above'] = 0
        state_edu_2020_map[state][age_group]['High school and above'] = 0
for name, val in edu_att_2020.items():
    col_label = name.split('!!')
    if len(col_label) > 1:
        state_name = col_label[0]
        if state_name in state_edu_2020_map.keys() and col_label[1] == 'Percent' and col_label[2] == 'Estimate':
            state_edu_2020_map[state_name]['25+']['Bachelor and above'] = percent(val[15])
            state_edu_2020_map[state_name]['25+']['High school and above'] = percent(val[14])
            state_edu_2020_map[state_name]['18-24']['Bachelor and above'] = percent(val[5])
            state_edu_2020_map[state_name]['18-24']['High school and above'] = round(sum([percent(i) for i in val[3:6]]), 1)
            state_edu_2020_map[state_name]['25-34']['Bachelor and above'] = percent(val[18])
            state_edu_2020_map[state_name]['25-34']['High school and above'] = percent(val[17])
            state_edu_2020_map[state_name]['35-44']['Bachelor and above'] = percent(val[21])
            state_edu_2020_map[state_name]['35-44']['High school and above'] = percent(val[20])
            state_edu_2020_map[state_name]['45-64']['Bachelor and above'] = percent(val[24])
            state_edu_2020_map[state_name]['45-64']['High school and above'] = percent(val[23])
            state_edu_2020_map[state_name]['65+']['Bachelor and above'] = percent(val[27])
            state_edu_2020_map[state_name]['65+']['High school and above'] = percent(val[26])
```

```python
with open('state-edu-data.json', 'w') as f:
    json.dump(state_edu_2020_map, f)
```

```python
state_poverty_2020_map = {}
fields = ['Poverty rate', 'Median household income']
for state in STATE_ABBR_MAP.keys():
    state_poverty_2020_map[state] = {}
    for field in fields:
        state_poverty_2020_map[state][field] = 0.0
```

```python
for _, row in saipe_2020.iterrows():
    if row[2] in STATE_ABBR_MAP.keys():
        state_poverty_2020_map[row[2]]['Poverty rate'] = float(row[6])
        state_poverty_2020_map[row[2]]['Median household income'] = int(row[21].replace(',', ''))
    else:
        print(row[2])
```

```python
with open('state-saipe-data.json', 'w') as f:
    json.dump(state_poverty_2020_map, f)
```

```python
popu_2020 = pd.read_csv('Population/population_by_race_2020.csv')
```

```python
popu_2020_map = {}
for name, item in popu_2020.items():
    if name in STATE_ABBR_MAP.keys():
        popu_2020_map[name] = {}
        popu_2020_map[name]['pop'] = int(item[0].replace(',', ''))
```

```python
homeless_2020 = pd.read_csv('AHAR/2007-2020-PIT-Estimates-by-state.csv')
```

```python
for _, row in homeless_2020.iterrows():
    if row[0] in ABBR_STATE_MAP.keys():
        state = popu_2020_map[ABBR_STATE_MAP[row[0]]]
        state['homeless'] = int(row[2].replace(',', ''))
        state['homeless_percent'] = round((state['homeless'] / state['pop']) * 100, 3)
```

```python
with open('state-pop-ahar-data.json', 'w') as f:
    json.dump(popu_2020_map, f)
```

```python
pls_data_min_max = {}
for col in rel_cols[1:]:
    minimum = float('inf')
    maximum = float('-inf')
    for year in range(1998, 2021):
        for state in STATE_ABBR_MAP.keys():
            if col in pls_data_map[year][state].keys():
                if pls_data_map[year][state][col] != -1 and pls_data_map[year][state][col] < minimum:
                    minimum = pls_data_map[year][state][col]
                if pls_data_map[year][state][col] > maximum:
                    maximum = pls_data_map[year][state][col]
    pls_data_min_max[col] = [minimum, maximum]
```

```python
with open('state-pls-sum-data-range.js', 'w') as f:
    json.dump(pls_data_min_max, f)
```
